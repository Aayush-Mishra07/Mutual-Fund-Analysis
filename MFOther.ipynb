## Mean centering the data:
Mean centering is a data preprocessing technique used to subtract the mean from each data point in a dataset. It helps to remove the average or central tendency from the data, making the mean of the centered data close to zero. This process is beneficial in various statistical analysis.This helps to compare any parameter or ratio of a scheme with the average of that parameter for all large cap schemes.

```

columns_to_center = data.columns[12:]

# Calculate the mean of each selected column
means = data[columns_to_center].mean()

# Subtract the mean from each selected column
df2= data[columns_to_center] - means
df2.sort_values(by='10 Years',ascending=False).head(10)
```

Intersection of 10Years,5 Years,3 Years,Sharpe values arranged from highest to lowest,
Expense Ratio values from lowest to highest:

```
teny=data.sort_values(by='10 Years',ascending=False).head(12)
fivey=data.sort_values(by='5 Years',ascending=False).head(12)
threey=data.sort_values(by='3 Years',ascending=False).head(12)
sharpe=data.sort_values(by='Sharpe',ascending=False).head(12)
exp=data.sort_values(by='Exp. Ratio',ascending=True).head(12)
top=threey.merge(teny, on='Scheme Name',how='inner').merge(fivey, on='Scheme Name',how='inner').merge(sharpe, on='Scheme Name',how='inner').merge(exp, on='Scheme Name',how='inner')
```

Common schemes that come in top 12 when arranged in descending order according to different values like 10 year,5 year,3 year returns,sharpe,exp ratio.
